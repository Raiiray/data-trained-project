```python
import sklearn
from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score,mean_squared_error , mean_absolute_error
from sklearn.model_selection import train_test_split
```

```python
x_train=df1.drop(["Price"],axis=1)
x_test=df2
y_train=df1["Price"]
```

```python
lr=LinearRegression()
lr.fit(x_train,y_train)
predlr=lr.predict(x_test)
print(lr.score(x_train,y_train))
```

```python
score=cross_val_score(lr,x_train,y_train,cv=5)
score.mean()
```

```python
dtr=DecisionTreeRegressor()
dtr.fit(x_train,y_train)
pred=dtr.predict(x_test)
print(dtr.score(x_train,y_train))
```

```python
score=cross_val_score(dtr,x_train,y_train,cv=5)
score.mean()
```

```python
rfr=RandomForestRegressor()
rfr.fit(x_train,y_train)
predrfr=rfr.predict(x_test)
print(rfr.score(x_train,y_train))
```

```python
score=cross_val_score(rfr,x_train,y_train,cv=5)
score.mean()
```

```python
knr=KNeighborsRegressor()
knr.fit(x_train,y_train)
predknr=knr.predict(x_test)
print(knr.score(x_train,y_train))
```

```python
score=cross_val_score(knr,x_train,y_train,cv=5)
score.mean()
```
